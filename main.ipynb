```python
import pandas as pd
df = pd.read_csv('input/steam.csv')
df.head()
```

```python
df.info()
df.isna().sum()
na_rows = df[df.isna().any(axis=1)]
print(na_rows)

df = df.dropna(subset=['publisher'])
display(df)

df.isna().sum()
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
df.info()
display(df)
```

```python
df['total_ratings'] = df['positive_ratings'] + df['negative_ratings']
most_ratings = df.sort_values('total_ratings', ascending=False)
df['positive_ratings_percent'] = df['positive_ratings']/df['total_ratings']*100
df_filtered = df[df['total_ratings'] > 1000]
best_ratings = df_filtered.sort_values('positive_ratings_percent', ascending=False)
display(most_ratings)
display(best_ratings)

most_ratings.to_csv('output/most_ratings.csv', index=False)
best_ratings.to_csv('output/best_ratings.csv', index=False)
```

```python
def get_price_range(price):
    if 0 < price < 10:
        return 'cheap'
    elif 10 <= price <= 30:
        return 'mid'
    elif price == 0:
        return 'free'
    else:
        return 'expensive'

def is_free(price):
    if price == 0:
        return True
    else:
        return False

df['price_range'] = df['price'].apply(get_price_range)
df['is_free'] = df['price'].apply(is_free)

filtered = df[df['total_ratings'] > 100]

mean_percent_by_price = (
    filtered
    .groupby('price_range')['positive_ratings_percent']
    .agg(['mean', 'size'])
)

mean_percent_by_free = (
    filtered
    .groupby('is_free')['positive_ratings_percent']
    .agg(['mean', 'size'])
)
display(mean_percent_by_price.reset_index())
display(mean_percent_by_free.reset_index())

mean_percent_by_price.reset_index().to_csv('output/mean_price.csv', index=False)
mean_percent_by_free.reset_index().to_csv('output/mean_free.csv', index=False)
```

```python

df_exploded = df.explode('genres').reset_index(drop=True)
display(df_exploded)

unique_genres = df_exploded['genres'].unique()
print(unique_genres)

genre_counts = (
    df_exploded
    .groupby('genres')['positive_ratings_percent']
    .agg(['mean', 'size'])
    .sort_values(by='mean', ascending=False)
    .reset_index()
)
genre_counts = genre_counts[(genre_counts['size'] > 250)]
print(genre_counts)

genre_counts.to_csv('output/genre_counts.csv', index=False)
```
